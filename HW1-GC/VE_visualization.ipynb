```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#from gc import garbage
names_list = ['Energy_load',	'Energy_distribution', 'Problem_Choice', 'Agent_choice','Resolutions', 'Oversights', 'Flights',	'Quickies',	'Prob_unsolved','Prob_shifts',	'Prob_active', 'Prob_latency','Dm_active', 'Dm_shifts', 'Energy_spent', 'Energy_excess', 'Choices_unsolved', 'Choices_active']
ve = pd. read_excel('garbage_can_v_41.xls', header = None, names = names_list)

#slice dataframe to get needed data
ve_data =  ve.loc[0:44 ,'Energy_distribution':'Flights' ]
```

```python
g = sns.FacetGrid(ve_data, col = 'Energy_distribution', row = 'Problem_Choice')
g.map_dataframe(sns.lineplot, 'Agent_choice','Resolutions')
```

```python
data_long = pd.melt(ve_data, ['Energy_distribution','Problem_Choice','Agent_choice'])
```

```python
g = sns.FacetGrid(data_long, col = 'Energy_distribution', row = 'Problem_Choice')
g.map_dataframe(sns.lineplot, x ='Agent_choice',y = 'value', hue = 'variable')
g.add_legend()
```
